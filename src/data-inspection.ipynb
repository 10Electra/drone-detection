```python
import sys, os, pathlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.animation as animation
from natsort import natsorted

left_dir  = pathlib.Path('../data/0/PADDED_RGB/')
right_dir = pathlib.Path('../data/0/Event/Frames/')
# left_dir  = '../data/0/Event/Frames/'
# right_dir = '../data/0/PADDED_RGB/'
# print(os.listdir(left_dir))
fps       = 24

def get_frames(folder):
    return natsorted(
        [f for f in folder.iterdir() if f.suffix.lower() in (".png", ".jpg", ".jpeg", ".tif", ".bmp")]
    )

left_frames  = get_frames(left_dir)
right_frames = get_frames(right_dir)
```

```python
img = plt.imread(left_frames[0])

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
for ax in axes:
    ax.axis("off")
plt.tight_layout(pad=0)

def update(i):
    for ax, img_path in zip(axes, (left_frames[i], right_frames[i])):
        ax.imshow(mpimg.imread(img_path))
    return axes

update(102)
```
